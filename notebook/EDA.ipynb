```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load data into DataFrame
data = pd.read_csv(r'C:\Users\salma elyagoubi\TSA\TSA\data\filtered_sensor.csv', sep=',')

# Merge 'BROKEN' and 'RECOVERING' statuses into a single 'BROKEN' category
data['machine_status'] = data['machine_status'].replace({'RECOVERING': 'BROKEN'})

# Calculate percentage of each machine status
status_counts = data['machine_status'].value_counts(normalize=True) * 100
print("Percentage of each machine status:")
print(status_counts)

# Define the selected sensors
selected_sensors = ['sensor_04', 'sensor_10', 'sensor_20']

# Function to identify anomalies
def identify_anomalies(sensor_data, threshold=3):
    mean = sensor_data.mean()
    std = sensor_data.std()
    anomalies = (sensor_data - mean).abs() > threshold * std
    return anomalies
```

```python
def calculate_sensor_means(data, sensors):
    """
    Calculate the mean for each sensor in the data.
    
    Parameters:
        data (pd.DataFrame): The input data frame containing sensor data.
        sensors (list): List of sensor column names.
        
    Returns:
        dict: Dictionary with sensor names as keys and their mean values as values.
    """
    means = {}
    for sensor in sensors:
        means[sensor] = data[sensor].mean()
    return means

def plot_anomalies(data, sensors, threshold=3):
    """
    Plot sensor data and highlight anomalies based on the mean and specified threshold.
    
    Parameters:
        data (pd.DataFrame): The input data frame containing sensor data and timestamps.
        sensors (list): List of sensor column names to plot.
        threshold (float): The threshold multiplier for detecting anomalies (default is 3).
    """
    fig, axs = plt.subplots(len(sensors), 1, figsize=(15, 5 * len(sensors)))

    for i, sensor in enumerate(sensors):
        sensor_data = data[sensor]
        mean_value = sensor_data.mean()
        std_dev = sensor_data.std()
        
        # Identify anomalies
        anomalies = (sensor_data - mean_value).abs() > threshold * std_dev
        
        # Plotting the sensor data
        axs[i].plot(data['timestamp'], sensor_data, label=sensor, color='blue')
        axs[i].scatter(data['timestamp'][anomalies], sensor_data[anomalies], color='red', label='Anomaly')
        
        axs[i].set_title(f'{sensor} with Anomalies')
        axs[i].set_xlabel('Timestamp')
        axs[i].set_ylabel('Value')
        axs[i].legend()
    
    plt.tight_layout()
    plt.show()
```
